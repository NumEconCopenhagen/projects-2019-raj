```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os
cwd = os.getcwd() #getting the working dictanary path.
print(cwd)
```

```python
infile= pd.read_stata('cphapts.dta') # imorting a stata file 

print(infile)
```

```python
#removing potential outliers
infile = infile[infile.price < 4000000]
infile= infile.drop('building_units', axis=1)
print(infile)

infile['age']=2019-infile['age']

infile['p_m2']=infile['price']/infile['m2']

#print(infile)
infile.hist(column='p_m2')
infile.hist(column='m2')
infile.hist(column='price')
infile.hist(column='age')
```

```python

from collections import Counter
a = infile['location']
letter_counts = Counter(a)
b = pd.DataFrame.from_dict(letter_counts, orient='index')
print(b)
#x = plt.barh(b, width=10)
b.plot(kind='bar')


plt.title('Apartments in CPH')
plt.xlabel('x')
plt.ylabel('y') 
```

```python
#using supervised learning
#heatmap code

import seaborn as sns
x = infile.corr()
#print(x)

ax=sns.heatmap(infile.corr(), square=True, cmap='RdYlGn',vmin=-1,vmax=1, linewidths=2) #the correlations between the variables in the data set
for label in ax.get_yticklabels():
    label.set_size(8)
for label in ax.get_xticklabels():
    label.set_size(8)
```
